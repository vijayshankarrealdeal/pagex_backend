```python
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI
from pydantic import BaseModel
import logging

# Load environment variables
load_dotenv()

# Define the expected output structure using Pydantic
class Query(BaseModel):
    user_query: str
    results: list[str]

# Setup logging to capture responses or errors
logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)

def run_query(query: str):
    try:
        # Initialize the model with the structured output
        llm = ChatGoogleGenerativeAI(
            model="gemini-2.5-pro-exp-03-25",
            temperature=0,
            max_tokens=500,  # Set a maximum token size to avoid truncation
            timeout=30,  # Allow more time for the model to process
            max_retries=2,
        ).with_structured_output(Query)

        # Log the query being passed
        logger.info(f"Sending query to model: {query}")

        # Invoke the model with the user query
        response = llm.invoke(query)

        # Log the raw response to see the result
        logger.info(f"Raw Response: {response}")

        if response:
            # If the model responds, return the structured output
            return response
        else:
            logger.warning("No response generated from the model.")
            return Query(user_query=query, results=["No results found"])

    except Exception as e:
        logger.error(f"Error invoking the model: {e}")
        return Query(user_query=query, results=[f"Error: {e}"])

# Example Query
query = "what is the capital of India"
result = run_query(query)

# Print the structured output
print(result)
```
